### 数据结构分析

```python
import json
# 从文件中读入. json.load方法
with open('.\data_dict.json','r',encoding='utf-8') as json_file:
    data_dict = json.load(json_file) 
```

```python
data_dict
```

```python
# json.loads方法将json字符串转化为dict
data_dict = json.loads(data_dict['data'])
```

```python
data_dict
```

```python
data_dict.keys()
```

```python
data_dict['foreignList'][0]
```

```python
# 字典元素处理。在countyr数据中，(如果有的话)弹出children这一项键值对，其余保留。
country_list =[]
children_list=[]
for country in data_dict['foreignList']:
    if country.get('children'):
        children=country.pop('children',None)
        children_list.append(children)
    country_list.append(country)    
```

```python
import pandas as pd
```

```python
# country_list是列表数据，直接转化为dataframe对象。
df =pd.DataFrame(country_list)
```
